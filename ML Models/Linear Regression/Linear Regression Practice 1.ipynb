# Problem 
You just got some contract work with an Ecommerce company based in New York City that sells clothing online but they also have in-store style and clothing advice sessions. Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want.

The company is trying to decide whether to focus their efforts on their mobile app experience or their website. They've hired you on contract to help them figure it out! 

```python
# Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading data and checking its headers

customers = pd.read_csv('Ecommerce Customers')
customers.head()
```

```python
# Checking out data info

customers.info()
```

```python
# Checking out data description

customers.describe()
```

```python
# Exploratory Data Analysis (Data visualization using seaborn)
# Finding relation between 'time on website' & 'yearly amount spent' columns 

#sns.set_palette("mako")
sns.set_style('whitegrid')
sns.jointplot(x='Time on Website', y='Yearly Amount Spent', data=customers)
```

```python
# Finding relation between 'time on App' & 'yearly amount spent' columns 

sns.jointplot(x='Time on App', y='Yearly Amount Spent', data=customers)
```

```python
# Finding relation between 'time on App' & 'length of membership' columns 

sns.jointplot(x='Time on App', y='Length of Membership', data=customers, kind='hex')
```

```python
# exploring column relationships across the entire data set

sns.pairplot(customers)
```

From the plot above, the 'Length of Membership' column looks more correlated with 'Yearly Amount Spent' column

```python
#Creating a linear model plot (using seaborn's lmplot) of 'Yearly Amount Spent' vs. 'Length of Membership'

sns.lmplot(x='Length of Membership', y='Yearly Amount Spent', data=customers)
```

```python
# dropping irrelevant columns for training and testing 

X = customers.drop(['Email','Address','Avatar','Yearly Amount Spent'],axis=1)
y = customers['Yearly Amount Spent']
```

```python
# splitting data into training and testing sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
# Training the model

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# predicting test data

predictions = lm.predict(X_test)
```

```python
# plotting a scatterplot of the real test values versus the predicted values

plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Evaluating the Linear model by calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error

from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test,predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
#plotting the resudial error (actual labels (y_test) - predicted labels(predictions))
sns.distplot((y_test-predictions),bins=50);
```

We see a normal distribution of the residual errors indicating a good model fit

```python
# creating a dataframe for the coefficients

coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

Interpreting the coefficients:
- Holding all other features fixed, **1 unit rise in the Avg. Session Length** lead to an increase of **25.91 dollars spent** 
- Holding all other features fixed, **1 unit rise in the Time on App** lead to an increase of **38.5 dollars spent** 
- Holding all other features fixed, **1 unit rise in the Time on Website** lead to an increase of **0.28 dollars spent** 
- Holding all other features fixed, **1 unit rise in the Length of Membership** lead to an increase of **61.16 dollars spent**
